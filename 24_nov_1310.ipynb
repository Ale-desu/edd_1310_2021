```
class nodoDoble:
  def __init__(self, valor, anterior=None,sig=None):
    self.dato=valor
    self.next=sig
    self.prev=anterior

class doubleLink:
  def __init__(self):
    self.__head=None
    self.__tail=None
    self.__size=0

  def get_size(self):
    return self.__size

  def is_empty(self):
    vrai=self.__head==None and self.__tail==None
    print(vrai)

  def append (self, val):
    if self.__head==None:
        nuevo=nodoDoble(val)
        self.__tail=self.__head=nuevo
    else:
        nuevo=nodoDoble(val,self.__tail,None)
        self.__tail.next=nuevo
        #nuevo.prev=self.__tail
        self.__tail= nuevo

  def transversal(self):
        curr_node=self.__head
        while curr_node != None:
            print(f"{curr_node.dato} ->",end="")
            curr_node=curr_node.next
        print("")

  def R_transversal(self):
        curr_node=self.__tail
        while curr_node != None:
            print(f"{curr_node.dato}  <-",end="")
            curr_node=curr_node.prev
        print("")

  def f_from_tail(self, val):
    cont=-1
    curr_node=self.__tail
    while curr_node!=None and curr_node.dato!=val:
      vrai=curr_node.dato==val
      curr_node=curr_node.prev
      cont=cont+1
    if curr_node.dato==val:
      print(f"el dato esta en la posicion:{cont}")
    else:
      print("el dato no existe")
      return cont

  def f_from_head(self, val):
    cont=-1
    curr_node=self.__head
    while curr_node!=None and curr_node.dato!=val:
      curr_node=curr_node.next
      cont=cont+1
    if curr_node.dato==val:
      print(f"el dato esta en la posicion:{cont}")
    else:
      print("el dato no existe")

  def R_from_head(self,value):
      curr_node=self.__head
      if self.__head.dato==value:
        self.__head=self.__head.next
      else:
          while curr_node.dato !=value and curr_node.next:
              curr_node=curr_node.next
              if curr_node.dato== value:
                  curr_node.prev.next=curr_node.next
                  curr_node.next.prev=curr_node.prev
                  curr_node.next=None
                  curr_node.prev=None
                  self.__size=-1
              else:
                  print("el dato no existe")

  def R_from_tail(self,value):
      curr_node=self.__tail
      if self.__tail.dato==value:
        self.__tail=self.__tail.prev
      else:
        while curr_node.dato !=value and curr_node.prev:
          curr_node=curr_node.prev
        if curr_node.dato== value:
          curr_node.prev.next=curr_node.next
          curr_node.next.prev=curr_node.prev
          curr_node.next=None
          curr_node.prev=None
          self.__size=-1
        else:
          print("el dato no existe")
```
